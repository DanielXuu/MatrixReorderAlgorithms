```python
with open('graphs/edges.txt', 'r') as f:
    l = f.read().splitlines()
```

```python
links = []
for i, e in enumerate(l):
    if i%3 == 0:
        links.append(e)
```

```python
ss = []
tt = []
for each in links:
    t = each.split('target')
    ss.append(t[0])
    tt.append(t[1])
```
